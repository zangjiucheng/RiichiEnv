```python
from riichienv import RiichiEnv
from riichienv.visualizer import GameViewer
from riichienv.agents import RandomAgent


agent = RandomAgent()
env = RiichiEnv(game_mode="4p-red-half")
obs_dict = env.reset()
while not env.done():
    actions = {player_id: agent.act(obs)
               for player_id, obs in obs_dict.items()}
    obs_dict = env.step(actions)

scores, ranks = env.scores(), env.ranks()
print(scores, ranks)

log_path = '../riichienv-ui/example_before_injection.jsonl'
viewer = GameViewer.from_jsonl(log_path, step=442, perspective=0, freeze=False)
viewer.show()
```

```python
summary = viewer.summary()
for round_summary in summary:
    print(round_summary)
```

```python
results = viewer.get_results(0)
for r in results:
    print(r)
    for yaku in r.yaku_list():
        print(f'  {yaku}')
```
